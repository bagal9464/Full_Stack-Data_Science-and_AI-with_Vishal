```python
feature_df = df[['empire','trade','god','coin']].drop_duplicates()
trade_counts =feature_df['trade'].fillna(0).astype(bool).sum()
religion_count = feature_df['god'].fillna(0).astype(bool).sum()
coin_count = feature_df['coin'].fillna(0).astype(bool).sum()
print(f"Trade Count: {trade_counts}, Religion Count: {religion_count}, Coin Count: {coin_count}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data for plotting
feature_names = ['Trade', 'Religion', 'Coin']
feature_values = [trade_counts, religion_count, coin_count]

# Plot
plt.figure(figsize=(6, 4))
sns.barplot(x=feature_names, y=feature_values, palette='Set2')

# Add title and labels
plt.title('Number of Empires with Each System')
plt.ylabel('Number of Empires')
plt.xlabel('System Type')

# Show values on top of bars
for i, value in enumerate(feature_values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
