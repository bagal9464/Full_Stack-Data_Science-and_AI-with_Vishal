# EDA Technique

```python
clean_data['Age']
```

```python
clean_data['Age']
```

```python
clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data['Age']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data['Exp']
```

```python
clean_data['Location']
```
