# Linear Regression Algorithum From Scratch

```python
# import nesesory ibraries
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# initialize data set
x = np.array([12,40,60,62,70,89])
y = np.array([24,44,66,59,75,98])
```

# Find the Mean of x independet and y dependent variable

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
print(f"x_mean = {x_mean}  y_mean = {y_mean}")
```

# Plot the actual data points with mean

```python
plt.figure(figsize = (6,3))
plt.scatter(x_mean,y_mean, color = 'red' ,label = 'Mean')
plt.scatter(x,y ,color = 'blue', label = 'actual data points')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title('Actual Data Points')
```

# Find the slope

```python
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean)**2)
```

```python
slope_m = numerator / denominator
slope_m
```

# Find Intercept

```python
intercept_c = y_mean - slope_m * x_mean
intercept_c
```

# Predict the values useing the Hypothesis function h(x)=β₀+β₁x

```python
y_hat = slope_m * x + intercept_c # y = mx + c
y_hat
```

```python
plt.figure(figsize = (6,3))
plt.plot(x,y_hat, color = 'red')
plt.scatter(x,y ,color = 'blue', label = 'actual data points')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.title('Actual Data Points')
```

# model evaluation in linear regression

# Find the Sum of Squered Error(SSE)

Which is tell us the difference between the actual data points and pridicted values  _
Ex.  Money “lost” (your model’s mistakes)

```python
SSE = np.sum((y - y_hat)**2)
print(SSE)
```

# Sum of Squered Regresser (SSR)

which is tell us How much of that spread my model explained _
EX.  Money “explained” by your model (how much you earned)

```python
SSR = np.sum((y_hat - y_mean)**2)
SSR
```

# Total Sum of Squere (SST)

How much spread is in the data overall  _ Ex.Total money (variation in data)

# Define Gradient Decent

```python


learning_rate = 0.000000000001
epochs = 100

for i in range(epochs):
    derivative_m = (-2/len(x))* np.sum(x* mse)
    derivative_c = (-2/len(x))* np.sum(mse)

    slope_m = slope_m - learning_rate * derivative_m
    intercept_c = intercept_c - learning_rate * derivative_c
    #print(slope_m , intercept_c)
    
print("Final parameters: ", slope_m, intercept_c) 
```

```python
y_hat = slope_m * x + intercept_c
y_hat
```
